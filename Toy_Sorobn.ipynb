Toy Bayes Net w/ Sorobn

```python
#installs
import sorobn as hh
import pandas as pd
import graphviz  
```

```python
df = hh.examples.asia().sample(500)
```

```python
test = hh.structure.chow_liu(df)
bn = hh.BayesNet(*test)
```

```python
dot = bn.graphviz()
dot
```

```python
path = dot.render('Toy', format='svg', cleanup=True)
```
